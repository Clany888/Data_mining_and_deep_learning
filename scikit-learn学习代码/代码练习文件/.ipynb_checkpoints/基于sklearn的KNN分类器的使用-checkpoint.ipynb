```python
from sklearn.datasets.samples_generator import make_blobs # 制作数据集
centers = [[-2,2],[3,5],[7,3]]  # 中心点
X,Y = make_blobs(n_samples=100,centers=centers,random_state = 0,cluster_std = 0.60)
# X样本点  Y自动生成的类别代号
# cluster_std标准差

```

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
plt.figure(figsize = (16,10),dpi = 100)
c = np.array(centers)
plt.scatter(X[:,0],X[:,1],c=Y,s=100,cmap='viridis')
plt.scatter(c[:,0],c[:,1],color = 'blue',s=100,marker="^")
# print(c)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
k=5
fen = KNeighborsClassifier(n_neighbors=k)
fen.fit(X,Y)

# 拟合模型
```

```python
yanben_x = [[0.6,4]]
yanben_y = fen.predict(yanben_x)
zuijin = fen.kneighbors(yanben_x,return_distance=False)  # return_distance=False不返回距离值
print(zuijin)
print(yanben_y)
```

```python
# 把上面的5个点标记出来
plt.figure(figsize = (16,10),dpi = 100)
c = np.array(centers)
plt.scatter(X[:,0],X[:,1],c=Y,s=100,cmap='viridis')
plt.scatter(c[:,0],c[:,1],color = 'blue',s=100,marker="^")
plt.scatter(yanben_x[0][0],yanben_x[0][1],color = 'red',s=100,marker="+")

for i in zuijin[0]:
    plt.plot([X[i][0], yanben_x[0][0]], [X[i][1], yanben_x[0][1]], 
             'k--', linewidth=0.6)


            
print([X[53][0], yanben_x[0][0]])
print([X[53][1], yanben_x[0][1]])
plt.show()
```

```python
jishu={'0':0,'1':0,"2":0}
aList = []
for j in zuijin[0]:
    aList.append(Y[j])
for x in aList:
    if x==0:
        jishu['0']+=1
    elif x==1:
        jishu['1']+=1
    else:
        jishu['2']+=1
print("输入的测试样本为:{}\n该样本属于类别:{}".format(yanben_x[0],max(jishu,key=jishu.get)))
```
