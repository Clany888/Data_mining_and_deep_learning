```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family'] = ['Arial Unicode MS', 'Microsoft Yahei', 'sans-serif'] #  全局设置支持中文字体，默认 sans-serif
```

# 1.生成数据集

```python
# 生成训练样本
n_dots = 200
X = 20 * np.random.rand(n_dots, 1)  
y = np.sin(X).ravel()  # ravel()进行降维操作，返回的是一个视图，会影响原始数据
# y = np.cos(X).ravel()  # ravel()进行降维操作，返回的是一个视图，会影响原始数据

# 添加一些噪声
y += 0.2 * np.random.rand(n_dots) - 0.1
print(np.shape(X),np.shape(y))
plt.figure(figsize=(12, 6)) 
plt.scatter(X[:,0],y[:])
plt.show()
```

# 2.构建模型

```python
from sklearn.neighbors import KNeighborsRegressor
k = 5
knn = KNeighborsRegressor(k)
knn.fit(X, y)
```

# 3.进行测试,计算泛化能力

```python
# 生成足够密集的点并进行预测
T = np.linspace(0, 20, 1000)[:, np.newaxis]  # np.newaxis 将一维数组转换为二维的 列
print(np.shape(T))
y_pred = knn.predict(T)  # 接收输入的数组类型测试样本，一般是二维数组，每一行是一个样本,每一列是一个属性
# predict 对于给定的500个测试样本,返回数组类型的预测结果

# 使用knn.score()方法计算拟合曲线针对训练样本集的拟合准确性
knn.score(X, y)
```

# 4.数据可视化,查看拟合的效果

```python
# 画出拟合曲线
plt.figure(figsize=(20, 12)) 
plt.scatter(X, y, c='g', label='训练样本', s=100)         # 画出训练样本

plt.plot(T, y_pred, c='k', label='测试样本集')  # 画出拟合曲线

plt.title("KNeighborsRegressor (k = %i)" % k)
plt.legend()
plt.show()
```
